```python
#TF_IDF ê³„ì‚°

from sklearn.feature_extraction.text import TfidfVectorizer

tfidf=TfidfVectorizer()

metadata1.reviewText=metadata1.reviewText.fillna('')

tfidf_matrix=tfidf.fit_transform(metadata1.reviewText)

tfidf_matrix.shape
```

```python
#TF_IDF ê³„ì‚° 2

from sklearn.metrics.pairwise import linear_kernel

cosine_sim = linear_kernel(tfidf_matrix,tfidf_matrix)
```

```python
#ì¶”ì²œ ë¦¬ìŠ¤íŠ¸ ìƒì„±



indices00=pd.Series(metadata1['songs_x']).drop_duplicates()
indices00=pd.Series(indices00.index, index=indices00).drop_duplicates()
#ë¦¬ë·°ë¥¼ ê¸°ë°˜ìœ¼ë¡œ ë…¸ëž˜ ì¶”ì²œ
def songs(songs_x,n=10,sim=cosine_sim):
    idx=indices00[songs_x]
    sim_sings=list(enumerate(cosine_sim[idx]))
    sim_sings=sorted( sim_sings, key= lambda x: x[1], reverse=True)
    sim_sings=sim_sings[1:(n+1)]
    
    sings_index=[i[0] for i in sim_sings]
    return indices00[sings_index]
    
    
#íŠ¹ì • ë…¸ëž˜ ê²€ìƒ‰
indices0=pd.Series(metadata1.index, index=metadata1['songs_x']).drop_duplicates()


def song(song,n=10):
    
    
    idx1=indices0[song]
    sel_song=list(enumerate(idx1))
    sel_song=sel_song[0:(n+1)]
    song_index=[i[1] for i in sel_song]
    
    return indices0[song_index]


#ê°€ìˆ˜ ê²€ìƒ‰

#ê°€ìˆ˜ ê²€ìƒ‰
indices1=pd.Series(metadata1.index, index=metadata1['artist']).drop_duplicates()


def singer(artist,n=10):
    
    
    idx2=indices1[artist]
    sel_artist=list(enumerate(idx2))
    sel_artist=sel_artist[0:(n+1)]
    singer_index=[i[1] for i in sel_artist]
    
    return metadata1.artist.iloc[singer_index]

# ìž¥ë¥´ì¶”ì²œ 
```
