# 추천 시스템 구현



```python
#유사도 계산(TF-IDF)

from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer()

metadata.artists=metadata.artists.fillna("")

tfidf_matrix=tfidf.fit_transform(metadata.artists)

tfidf_matrix.shape
```

```python
#sklearn의 cosine 유사도 사용

from sklearn.metrics.pairwise import linear_kernel

cosine_sim = linear_kernel(tfidf_matrix,tfidf_matrix)
```

```python



#가수 검색
indices0=pd.Series(metadata1.index, index=metadata1['artists']).drop_duplicates()

indices00=pd.Series(metadata1['artists']).drop_duplicates()
indices00=pd.Series(indices00.index, index=indices00).drop_duplicates()
def singer(artists,n=10):
    
    
    idx=indices0[artists]
    sel_artist=list(enumerate(idx))
    sel_artist=sel_artist[0:(n+1)]
    singer_index=[i[1] for i in sel_artist]
    
    return metadata1.artists.iloc[singer_index]

#가수 추천
def artist(artists, n=10,sim=cosine_sim):
    
    idx1=indices00[artists]
    sim_artist=list(enumerate(cosine_sim[idx1]))
    sim_artist=sorted(sim_artist, key= lambda x: x[1], reverse=True )
    sim_artist= sim_artist[1:(n+1)]
    
    artist_index=[i[0] for i in sim_artist]
    
    return indices00[artist_index]
    #return metadata1.artists.iloc[artist_index].drop_duplicates()


#속도 + 에너지    몇번째를 가져올것인지 입력
indices2=pd.Series(metadata1.index, index=metadata1['energy']).drop_duplicates()
indices1=pd.Series(metadata1['tempo']).drop_duplicates()


    
    
indices1=round(indices1,4)
def tempo(tempo,n=10):
    
    idx=indices1[tempo]
    tempo_s=np.subtract(indices1,idx)
    #tempos_s = round(tempos_s,4)
    tempo_s=abs(tempo_s)
    tempo_s=tempo_s.sort_values(ascending=False)
    #tempo_s=sorted(tempo_s, key=lambda x: x[1], reverse=True)
   
    tempo_s=tempo_s[1:(n+1)]
   
   
    
    #tempo_index=[i for i in tempo_s]
    #return metadata1.tempo.iloc[tempo_s]
    return tempo_s
```
