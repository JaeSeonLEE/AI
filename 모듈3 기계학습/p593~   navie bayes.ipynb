```python
#p601 navie bayes  텍스트로 벡터 만들기

from numpy import *

def loadDataSet():
    postingList=[['my','dog','has','flea','problems','help','please'],
                 ['maybe','not','take','him','to','dog','park','stupid'],
                 ['my','dalmation','is','so','cute','l','love','him'],
                 ['stop','posting','stupid','worthless','garbage'],
                 ['mr','licks','ate','my','steak','how','to','stop','him'],
                 ['quit','buying','worthless','dog','fodd','stupid']]
    
    classVec=[0,1,0,1,0,1]   # 1 폭력적 , 0 비폭력적
    
    return postingList,classVec

def createVocabList(dataSet):
    vocabSet=set([])    #각 문서로부터 새로운 집합의 변수 생성하여 첨부
    
    for document in dataSet:
        vocabSet=vocabSet | set(document)
        
    return list(vocabSet)

def set0fWords2Vec(vocabList,inputSet):
    returnVec=[0]*len(vocabList)
    for word in inputSet:
        if word in vocabList:
            returnVec[vocabList.index(word)]=1
        
        else:print("the word: {} is not in my vocabulary!".format(word))
     
    return returnVec


list0Posts,listClasses=loadDataSet()


myVocabList=createVocabList(list0Posts)



set0fWords2Vec(myVocabList, list0Posts[0])

#p603 훈련: 단어 벡터로 확률 계산하기

def trainNBO(trainMatrix,trainCategory):
    numTrainDocs=len(trainMatrix)
    numWords=len(trainMatrix[0])
    pAbusive=sum(trainCategory)/float(numTrainDocs)
    p0Num=ones(numWords); p1Num=ones(numWords)
    p0Denom=2.0 ; p1Denom=2.0
    for i in range(numTrainDocs):
        if trainCategory[i]==1:
            p1Num += trainMatrix[i]
            p1Denom += sum(trainMatrix[i])
         
        else:
            p0Num += trainMatrix[i]
            p0Denom += sum(trainMatrix[i])
            
    p1Vect= log(p1Num/p1Denom)
    p0Vect= log(p0Num/p0Denom)
    
    return p0Vect, p1Vect ,pAbusive


trainMat=[]
for postinDoc in list0Posts:
    trainMat.append(set0fWords2Vec(myVocabList,postinDoc))

p0V,p1V,pAb = trainNBO(trainMat, listClasses)

#p604 실제 조건을 반영 하기 위해 분류기 수정

def classifyNB(vec2Classify,p0Vec,p1Vec,pClass1):
    p1=sum(vec2Classify * p1Vec)+ log(pClass1)
    p0=sum(vec2Classify * p0Vec)+ log(1.0- pClass1)
    
    if p1> p0:
        return 1
    
    else:
        return 0
    
def testingNB():
    list0Posts,listClasses= loadDataSet()
    testEntry=['love','my','dalmation','jinman']
    thisDoc=array(set0fWords2Vec(myVocabList,testEntry))
    print("{} classified as : {} ".format(testEntry,classifyNB(thisDoc,p0V,p1V,pAb)))
    
    testEntry=['stupid','garbage']
    
    thisDoc=array(set0fWords2Vec(myVocabList,testEntry))
    print(" {} clssified as: {} ".format(testEntry,classifyNB(thisDoc,p0V,p1V,pAb)))
```

```python

testingNB()
```
