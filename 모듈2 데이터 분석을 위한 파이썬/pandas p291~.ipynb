```python
# p321  데이터 합치기


from pandas import Series,DataFrame
from numpy import nan as NA
import pandas as pd
import numpy as np

s1=Series([0,2,3],index=["a","b","c"])
s2=Series([2,3,4,5],index=["a","b","d","e"])
s3=Series([1,3],index=["e","f"])

print("s1 \n{0}  \n s2 \n{1} \n s3 \n {2}".format(s1,s2,s3))
print("\n---------------------\n")
print("s1~s3 행 이어 붙이기 \n {0}". format(pd.concat([s1,s2,s3])))
print("\n---------------------\n")
print("s1~s3 열 이어 붙이기 \n{0}". format(pd.concat([s1,s2,s3],axis=1,sort=True)))

pd.concat([s1,s2],axis=1,join="inner",sort=True)

```

```python
# p327

from pandas import Series,DataFrame
from numpy import nan as NA
import pandas as pd
import numpy as np

df=DataFrame({"name":["kim","lee","bae","cho","kang"],"sex":[1,2,2,1,1],"home":["마포","수지","처인","용산","종로"],"income":[250,195,325,210,274]})
num_to_str={1:"남", 2:"여"}

to_city={"마포":"서울","용산":"서울","종로":"서울","수지":"용인","처인":"용인"}

def f(x):
    return x*2.015

df["gender"]=df["sex"].map(num_to_str)
df["city"]=df["home"].map(to_city)
df["remium"]=df["income"].map(f)
```
