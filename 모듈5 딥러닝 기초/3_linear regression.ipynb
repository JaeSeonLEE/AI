```python
import numpy as np
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
import tensorflow as tf

gpu_options=tf.GPUOptions(per_process_gpu_memory_fraction=0.333)
sess=tf.Session(config=tf.ConfigProto(gpu_options=gpu_options))
x_train = np.array([1, 2, 3])
y_train = np.array([1, 2, 3])

model = Sequential()
model.add(Dense(1, input_dim=1))                  #1 노드 수 

sgd = SGD(lr=0.1)                            #SGD : 경사 하강법 중 하나 , Ir= 학습률
model.compile(loss='mse', optimizer=sgd)          #compile :학습 지정  , loss= 오차  

# prints summary of the model to the terminal
model.summary()

model.fit(x_train, y_train, epochs=200)                 # 데이터 , 정답, 실해수

y_predict = model.predict(np.array([5]))
print(y_predict)
```

```python
import numpy as np
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential


x_data=np.array([[73,80,75],
                 [93,88,93],
                 [89,91,90],
                 [96,98,100],
                 [73,66,70]])

y_data=np.array([152,185,180,196,142])





model = Sequential()
model.add(Dense(1, input_dim=3))                  #1 노드 수 

sgd = SGD(lr=0.000001)                            #SGD : 경사 하강법 중 하나 , Ir= 학습률
model.compile(loss='mse', optimizer=sgd)          #compile :학습 지정  , loss= 오차  

# prints summary of the model to the terminal
model.summary()

model.fit(x_data, y_data, epochs=200)                 # 데이터 , 정답, 실해수
test=np.array([73,80,75])
y_predict = model.predict(np.expand_dims(test,axis=0))         #test 차원 확대하여 예측
print(y_predict)

```
