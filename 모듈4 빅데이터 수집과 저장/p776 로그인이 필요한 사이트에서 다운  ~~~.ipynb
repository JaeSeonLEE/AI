# 예제  네이버 API 활용

```python
import os
import sys
import datetime
import time
import urllib.request
import json



#접속 
def get_request_url(url):
    client_id="rGDXdVj4npokHuNwqhON"
    client_secrret="ms9fxWzkJb"
    rep=urllib.request.Request(url)
    rep.add_header("X-Naver-Client-Id",client_id)     #X-Naver-Client-Id 변수명은 바꾸면 안됨
    rep.add_header("X-Naver-Client-Secret",client_secrret)
    
    try:
        response=urllib.request.urlopen(rep)
        if response.getcode() == 200 :    # 정상처리
            print("[%s] Url request success" % datetime.datetime.now() )
            return response.read().decode("utf-8")
        
    except Exception as e:               #에러
        print(e)
        print("[%s] Error for URL : %s" % (datetime.datetime.now(),url))
        return None    
    
    #검색결과 받아오기
    
def getNaverSearchResult(sNode,search_text, page_start, display):    #카테고리(뉴스, 카페등)검색어   시작 페이지 표시될수
    base="https://openapi.naver.com/v1/search"            #검색어에서 실행할 위치
        
    node="/%s.json" % sNode
    params="?query=%s&start=%s&display=%s" % (urllib.parse.quote(search_text),
                                                    page_start,display)
        
    url=base+node+params
        
    getData=get_request_url(url)
        
    if (getData == None):                          # 받은 데이터가 없을 경우
            return None
        
    else:                                            # 받은 데이터가 있으면 json 형태로 반환
         return json.loads(getData)
    
#비로그인 방식으로 네이버 검색
def getPostData(post, jsonResult):
    title=post["title"]
    description=post["description"]
    link=post["link"]
    org_link=post["originallink"]
    
    pDate=datetime.datetime.strptime(post['pubDate'],
                                            "%a, %d %b %Y %H:%M:%S +0900")
    
    pDate=pDate.strftime('%Y-%m-%d %H:%M:%S')
    
    jsonResult.append({"title":title, "description":description, 
                       "org_link":org_link, "link":link, "pDate":pDate})
    
    return


# 매인

def main():
    jsonResult=[]
    sNode="news"   #blog,cafearticle
    search_text="파이썬"
    display_count=100                   #변경해도 상관 없음
    page_start=1
    
    jsonSearch= getNaverSearchResult(sNode,search_text, page_start, display_count)
    
    #필요한 데이터 나눠서 저장
    while ((jsonSearch != None) and (jsonSearch["display"] != 0)) :
        for post in jsonSearch["items"]:
            getPostData(post, jsonResult)
        
        nStart=jsonSearch["start"]+jsonSearch["display"]                # nStart:다음페이지 시작번호
        
        jsonSearch=getNaverSearchResult(sNode,search_text,nStart , display_count)
        
    with open('%s_naver_%s.json' % (search_text, sNode), "w", encoding="utf-8") as outfile:
        retJson=json.dumps(jsonResult,indent=4,sort_keys=True,ensure_ascii=False )    #가독성을 높이기위해 들여쓰기
                                                                                    #한글을 위해 아스키코드 처리no
        outfile.write(retJson)
    
    print('%s_naver_%s.json SAVED' % (search_text,sNode))
    
if __name__== '__main__':
    main()
```
